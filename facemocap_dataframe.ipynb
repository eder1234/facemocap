```python
import pickle
import pandas as pd

# Load control_df and patients_df
with open('control_df.pkl', 'rb') as f:
    control_df = pickle.load(f)
with open('patients_df.pkl', 'rb') as f:
    patients_df = pickle.load(f)

# Concatenate control_df and patients_df
facemocap_df = pd.concat([control_df, patients_df], axis=0)
facemocap_df.sample(50)
```

```python
import numpy as np
# Verify nan value in facemocap_df["Date"][401]
np.isnan(facemocap_df["Date"][401])
```

```python
for index, row in facemocap_df.iterrows():
    if row["Date"] == "":
        facemocap_df["Date"][index] = np.nan
print(facemocap_df["Date"][401])
```

```python
facemocap_df.sample(50)
```

```python
facemocap_df.to_pickle("facemocap_df.pkl")
```
